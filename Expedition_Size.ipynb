```python
import pandas as pd

from google.colab import files
uploaded = files.upload()


filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_filtered = df[df['Year'] >= 1990]

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Year', y='expedition_size', data=df_filtered)
plt.title('Expedition Size Over the Years')
plt.xlabel('Year')
plt.ylabel('Expedition Size')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

ycol = "expedition_size"

# Drop rows where 'expedition_size' is NaN, and filter for years after 1990
df_cleaned = df.dropna(subset=[ycol, 'Year'])
df_cleaned = df_cleaned[df_cleaned['Year'] > 1990]

plt.figure(figsize=(14,6))
plt.hist2d(df_cleaned["Year"], df_cleaned[ycol], bins=[36, 50], cmap='BuPu')  # Changed colormap to 'viridis'
plt.title("Expedition Size vs Year (Density) (After 1990)")
plt.xlabel("Year")
plt.ylabel("Expedition Size")
plt.colorbar(label="Count")
plt.grid(False)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

ycol = "expedition_size"

# Filter data for years after 1990
df_filtered_post_1990 = df[df['Year'] > 1990]

x = df_filtered_post_1990["Year"].to_numpy()
y = df_filtered_post_1990[ycol].to_numpy()

# jitter the x slightly so points dont stack exactly
x_jitter = x + np.random.uniform(-0.15, 0.15, size=len(x))

plt.figure(figsize=(14,6))
plt.scatter(x_jitter, y, alpha=0.25, s=15)
plt.title("Expedition Size Over the Years (Jittered Scatter, After 1990)")
plt.xlabel("Year")
plt.ylabel("Expedition Size")
plt.grid(True, alpha=0.3)
plt.show()
```
